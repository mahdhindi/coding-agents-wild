```python
import pandas as pd

# 1) Direct URL to the AIDev pr_commits table (this is the AIDev-pop commit table)
PR_COMMITS_URL = "https://huggingface.co/datasets/hao-li/AIDev/resolve/main/pr_commits.parquet"

# 2) Load the parquet file into a DataFrame
print("Loading pr_commits.parquet from Hugging Face...")
df_commits = pd.read_parquet(PR_COMMITS_URL)

print("Shape of commits table:", df_commits.shape)
print(df_commits.head())

# 3) Save as CSV
OUTPUT_CSV = "aidev_pop_pr_commits.csv"
df_commits.to_csv(OUTPUT_CSV, index=False)
print(f"Saved to {OUTPUT_CSV}")
```
